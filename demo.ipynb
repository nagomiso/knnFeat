```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x0 = np.random.rand(500) - 0.5
x1 = np.random.rand(500) - 0.5
X = np.array(list(zip(x0, x1)))
y = np.array([1 if i0 * i1 > 0 else 0 for (i0, i1)  in list(zip(x0, x1))])
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
from knn_feat import knnExtract
```

```python
%%time
newX = knnExtract(X, y, k=1, holds = 5)
```

```python
plt.scatter(newX[:,0], newX[:,1], c=y)
```

```python
from sklearn_api import KNeighborsFeatures
```

```python
knf_faiss = KNeighborsFeatures(method='faiss', n_jobs=-1, random_state=42)
```

```python
%%time
new_X =knf_faiss.fit_transform(X=X, y=y)
```

```python
plt.scatter(new_X[:,0], new_X[:,1], c=y)
```

```python
knf_skl = KNeighborsFeatures(method='sklearn', n_jobs=-1, random_state=42)
```

```python
%%time
new_X =knf_skl.fit_transform(X=X, y=y)
```

```python
plt.scatter(new_X[:,0], new_X[:,1], c=y)
```
